```python
import torch
```



* arange(n) 으로 0,1 ... n-1 값을 갖는 새로운 텐서를 만들 수 있다. 디폴트 간격은 1이고, 달리 지정하지 않는 한 새 텐서는 메인 메모리에 저장되며 CPU 기반 계산을 위해 지정됨.



```python
x = torch.arange(12, dtype=torch.float32)
x
```

* shape 으로 텐서의 shape 알 수 있고, numel()로 전체 element 개수를 알 수 있음. reshape 이용하여 shape 변경 가능(텐서가 자동으로 유추하기를 원하는 차원에 대해 -1을 배치할 수 있다)

```python
x.shape
```

```python
x.numel()
```

```python
X = x.reshape(3, 4) # x.reshape(-1, 4) 또는 x.reshape(3, -1) 도 가능 
X
```

* zeros, ones 이용하여 0,1로 초기화된 텐서 선언 가능

```python
torch.zeros((2, 3, 4))
```

```python
torch.ones((2, 3, 4))
```

* randn 이용하여 0 ~ 1 사이 랜덤 값으로 초기화 가능

```python
torch.randn(3, 4)
```

* 특정 값으로 지정하여 초기화 가능

```python
torch.tensor([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
```
